```python
a = 15
b = 25
c = 30

if(b<a>c):
  print('a is the greatest number')
elif(a<b>c):
  print('b is the greatest number')
else:
  print('c is the greatest number')
```

```python
a = 20
b = 40
c = 60

if(a>b):
  if(a>c):
    print('a is greatest number')
  else:
    print('c is the greatest number')
else:
  if(b>c):
    print('b is the greatest number')
  else:
    print('c is the greatest number')
```

#Function

Fuction is a block of code that can be reused in a program

factorial of number

Factorial of a number is the product of all the positive integers less than or equal to the givin number



Factorial of 5 = 5 x 4 x 3 x 2 x 1 = 120

```python
number = int(input('Enter a number to find its factorial: '))

factorial = 1

if number == 0:
  print('the factorial of 0 is 1')
else:
  for i in range(1,number + 1):  #range of (1,6) --> 1,2,3,4,5
    factorial = factorial * i
  print(factorial)
```

facttorial function

```python
def factorial_value(num):

  factorial = 1

  if num == 0:
    return factorial

  else:
    for i in range(1,num+1):
      factorial = factorial*i
    return factorial
```

```python
print(factorial_value(5))
```

```python
print(factorial_value(10))
```
