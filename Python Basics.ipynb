Arithmetic Operators

```python
num_1 = 20
num_2 = 10

#addition
sum = num_1 + num_2
print(sum)

#subtraction
sub = num_1 - num_2
print(sub)

#Multiplication
mul = num_1 * num_2
print(mul)

#division
div = num_1 / num_2
print(div)

#exponent

exp = num_1 ** num_2
print(exp) #20^10

#modulus

mod = num_1 % num_2
print(mod)
```

Comparision Opertors

```python
a = 5
b = 10

print(a == b)  #equal to
print(a != b)  #not equal to
print(a > b)   #greater than
print(a < b)
print(a >= b)
print(a <= b)
```

```python
a = 5
b = 10

c = [1,2,3,4,5]

print(a in c)
print(b in c)
print(a not in c)
print(b not in c)
```
