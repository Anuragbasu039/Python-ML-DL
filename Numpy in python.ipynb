List Vs Numpy - Time Taken

```python
from time import process_time
```

Time taken by a list

```python
python_list = [i for i in range(10000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5  # np_array = np_array + 5

end_time = process_time()

print(end_time - start_time)
```

Initial Placeholders in numpy arrays



```python
# create a numpy array of zeros

x = np.zeros((4,5))
print(x)
```

```python
# create a numpy array of ones

y = np.ones((3,3))
print(y)
```

```python
# create an array of particular value

z = np.full((5,4),5)
print(z)
```

```python
# create an identity matrix

a = np.eye(4)
print(a)
```

```python
# create a numpy array with random values

b = np.random.random((3,4))
print(b)
```

```python
# random integer value array within a specific range

c = np.random.randint(10, 100,(3, 5))
print(c)
```

```python
# array of evenly spaced value --> specifying the number of value required

d = np.linspace(10,30,5)
print(d)
```

```python
# array of evenly spaced value --> specifying the step

e = np.arange(10,30,5)
print(e)
```

```python
# convert a list to a numpy array

list2 = [10,20,30,40,50]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

Mathamatical operation on a numpy array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

print(list1 + list2)  # conatenate or join two array
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
print(a)
print(b)
```

```python
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

Array manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# tranpose

trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# reshaping a array

a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
